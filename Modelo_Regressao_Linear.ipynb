```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
```

Criando a pool de dados aleatorios atraves da lib matplot

```python
x,y = make_regression(n_samples=150, n_features=1, noise=30)
plt.scatter(x,y)
```

Aqui treinamos os modelos com os dados.

```python
modelo = LinearRegression()
modelo.fit(x,y)
```

No cod abaixo definimos os coeficientes angurales e lineares para nossa regressao

```python
a_coeff = modelo.coef_
l_coeff = modelo.intercept_

print('Coeficiente Angular: {0}\nCoeficiente Linear: {1}'.format(a_coeff,l_coeff))
```

Com base nos dados plotamos a nossa Regressao Linear e podemos prever valores X (linha 3 de exemplo)

```python
plt.scatter(x,y)
plt.plot(x, l_coeff + a_coeff*x, color='red')
plt.scatter(1.8796, l_coeff + a_coeff*1.8796, color='green')
plt.show()
```
