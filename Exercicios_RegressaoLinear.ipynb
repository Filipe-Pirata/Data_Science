```python
import numpy as np
from matplotlib import pyplot as plt


x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 2, 3, 4, 5])

w = 0.5
b = 0.5

plt.scatter(x, y, marker='x', color='red', label="Dados reais")
plt.plot(x, w * x + b, label="Predi칞칚o Inicial", color="blue")
plt.xlabel("Popula칞칚o")
plt.ylabel("Lucro")
plt.legend()
plt.show()


#Parametros do Gradient
theta_0 = 0 # Intercepto
theta_1  = 0 # Coeficiente angular
alpha = 0.01 # valor da taxa de aprendizado
iterations = 1000
m = len(x)

# Gradient Descendente
for _ in range(iterations):
  h = theta_0 + theta_1 * x  # Hipotese
  erro = h - y #Previsao - Real

  # Calculo do Gradient

  grad_theta_0 = (1/m) * np.sum(erro)
  grad_theta_1  = (1/m) * np.sum(erro * x)

  theta_0 -= alpha * grad_theta_0
  theta_1 -= alpha * grad_theta_1

# Par칙metros  aprendidos
print(f"Par칙metros finais: theta_0 = {theta_0:.4f}, theta_1 = {theta_1:.4f}")

#Funcao de Custo

def compute_cost(x, y, w, b):
    m = x.shape[0] # numero de exemplos
    cost_sum = 0

    for i in range(m):
        f_wb = w * x[i] + b
        cost = (f_wb - y[i]) ** 2
        cost_sum += cost
    total_cost = (1 / (2 * m)) * cost_sum
    return total_cost

custo = compute_cost(x, y, theta_1, theta_0)
print(f"Custo do modelo: {custo:.2f}")

print("###### NOVOS DADOS ######")


x_novos = np.array([6, 7, 8, 9, 10])
y_previstos = theta_0 + theta_1 * x_novos

plt.scatter(x, y, marker='x', color='red', label="Dados reais")
plt.plot(x, theta_1 * x + theta_0, label="Predi칞칚o Final", color="green")

for i in range(len(x_novos)):
  print(f"Para x = {x_novos[i]}, previs칚o de y = {y_previstos[i]:.2f}")

```
